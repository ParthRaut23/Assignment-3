```python
#Q3) Compare the average PM2.5 levels during weekends and weekdays for the station Lal Bahadur Shastri Nagar, Kalaburagi - KSPCB over the year 2021. 
# Create a plot to show these trends Steps:
# Calculate the monthly average PM2.5 for weekdays and weekends separately
# X-axis: Months (January to December).
# Y-axis: Average PM2.5 levels.
# Use two lines: one for the monthly average PM2.5 levels on weekdays and one for weekends.



station_name = "Lal Bahadur Shastri Nagar, Kalaburagi - KSPCB"

# Ensure Timestamp is in datetime format
data["Timestamp"] = pd.to_datetime(data["Timestamp"])

# Filter data for 2021 and specific station
data_2021 = data.loc[
    (data["Timestamp"].dt.year == 2021) & (data["station"] == station_name)
].copy()

# Extract weekday (0=Monday, 6=Sunday)
data_2021["Weekday"] = data_2021["Timestamp"].dt.weekday  

# Create a weekend column (True for Sat-Sun)
data_2021["Is_Weekend"] = data_2021["Weekday"] >= 5  

# Calculate monthly averages
monthly_weekday_avg = (
    data_2021[~data_2021["Is_Weekend"]]
    .groupby(data_2021["Timestamp"].dt.month)["PM2.5"]
    .mean()
)
monthly_weekend_avg = (
    data_2021[data_2021["Is_Weekend"]]
    .groupby(data_2021["Timestamp"].dt.month)["PM2.5"]
    .mean()
)

# Print results
print(monthly_weekend_avg)
print()

# Plot the trends
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(monthly_weekday_avg.index, monthly_weekday_avg.values, label="Weekdays", marker="o")
plt.plot(monthly_weekend_avg.index, monthly_weekend_avg.values, label="Weekends", marker="s", linestyle="--")

plt.xlabel("Month")
plt.ylabel("Average PM2.5 (µg/m³)")
plt.title("Monthly PM2.5 Trends (Weekdays vs Weekends) - 2021")
plt.xticks(range(1, 13), ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.legend()
plt.grid()
plt.show()
```

```python
#Q1) For each state, calculate the percentage increase or decrease in PM2.5 levels during monsoon months (June to September) compared to 
# summer months (March to May). Which state has the most difference (increase or decrease) for the year 2022?


# Filter data for 2022
data_2022 = data[data["Timestamp"].dt.year == 2022]

# Define seasons
monsoon_months = [6, 7, 8, 9]
summer_months = [3, 4, 5]

# Group data by state and calculate seasonal PM2.5 averages
state_pm25_monsoon = data_2022[data_2022["Timestamp"].dt.month.isin(monsoon_months)].groupby("state")["PM2.5"].mean()
state_pm25_summer = data_2022[data_2022["Timestamp"].dt.month.isin(summer_months)].groupby("state")["PM2.5"].mean()
print("The PM2.5 average data by state for monsoon season is:\n",state_pm25_monsoon)
print()
print("The PM2.5 average data by state for summer season is:\n",state_pm25_summer)
print()

# Calculate percentage change
percentage_change = ((state_pm25_monsoon - state_pm25_summer) / state_pm25_summer) * 100
print("The PM2.5 percentage change for each state between summer and monsoon is:\n", percentage_change)

# Find the state with the highest change
most_changed_state = percentage_change.abs().idxmax()
most_changed_value = percentage_change[most_changed_state]

print(f"\nThe state with the most significant PM2.5 change between summer and monsoon in 2022 is {most_changed_state} with a {most_changed_value:.2f}% change.")
```

```python
#Q2) Plot the average PM2.5 levels for Delhi across each season (Winter: December-February, Summer: March-May, Monsoon: June-September) from 2017 to 2023. 
# What seasonal variation, if any, can you observe in pollution levels?


# Filter for Delhi
delhi_data = data[data["state"] == "Delhi"]

# Define seasons
seasons = {
    "Winter": [12, 1, 2],
    "Summer": [3, 4, 5],
    "Monsoon": [6, 7, 8, 9]
}

# Calculate seasonal PM2.5 averages per year
seasonal_trends = []
for year in range(2017, 2024):
    year_data = delhi_data[delhi_data["Timestamp"].dt.year == year]
    for season, months in seasons.items():
        avg_pm25 = year_data[year_data["Timestamp"].dt.month.isin(months)]["PM2.5"].mean()
        seasonal_trends.append({"Year": year, "Season": season, "PM2.5": avg_pm25})

# Convert to DataFrame
seasonal_df = pd.DataFrame(seasonal_trends)
display(seasonal_df)

# Plot the trends
plt.figure(figsize=(10, 6))
sns.barplot(data=seasonal_df, x="Year", y="PM2.5", hue="Season")

plt.xlabel("Year")
plt.ylabel("Average PM2.5 (µg/m³)")
plt.title("Seasonal PM2.5 Levels in Delhi (2017-2023)")
plt.legend(title="Season")
plt.xticks(rotation=45)
plt.grid()
plt.show()


# We can observe the following seasonal variation in pollution levels:
# Winter has the highest PM2.5 levels, followed by summer, while monsoon has the lowest. This is due to winter inversions trapping pollutants, 
# summer winds dispersing them, and monsoon rains cleaning the air.
```

```python
#Q1) Compare the average PM2.5 levels between states that received NCAP funding and those that did not in 2021.



# Extract relevant funding data
ncap_funding_2021 = ncap_funding[["state", "Total fund released"]]

# Convert funding to numeric (handling missing values)
# Convert funding to numeric safely using .loc
ncap_funding_2021.loc[:, "Total fund released"] = pd.to_numeric(ncap_funding_2021["Total fund released"], errors="coerce")


# Identify funded states (non-null total fund released)
funded_states = ncap_funding_2021.dropna()["state"].unique()

# Add funding status to data
data["Funding_Status"] = data["state"].apply(lambda x: "Funded" if x in funded_states else "Non-Funded")

# Filter data for 2021
data["Timestamp"] = pd.to_datetime(data["Timestamp"])
data_2021 = data[data["Timestamp"].dt.year == 2021]

# Compute average PM2.5 for each group
avg_pm25_funding = data_2021.groupby("Funding_Status")["PM2.5"].mean()

print(avg_pm25_funding)

# Bar plot for funded vs. non-funded states
plt.figure(figsize=(6, 4))
sns.barplot(x=avg_pm25_funding.index, y=avg_pm25_funding.values, hue = avg_pm25_funding.index, palette="coolwarm", legend = False)
plt.xlabel("Funding Status")
plt.ylabel("Average PM2.5 (2021)")
plt.title("Comparison of PM2.5 Levels: Funded vs. Non-Funded States (2021)")
plt.show()
```

```python
#Q2) Create a time series plot showing PM2.5 levels with an increment in NCAP funding for Assam. Has an increment of the financing led to noticeable 
# improvements in air quality?



# Convert funding columns to numeric, replacing '-' with NaN
funding_columns = ["Amount released during FY 2019-20", 
                   "Amount released during FY 2020-21", 
                   "Amount released during FY 2021-22", 
                   "Total fund released"]

for col in funding_columns:
    ncap_funding[col] = pd.to_numeric(ncap_funding[col], errors="coerce")

# Aggregate total funding per state
state_funding = ncap_funding.groupby("state")[funding_columns].sum().reset_index()

# Reshape funding data to have a "Year" column
state_funding_melted = state_funding.melt(id_vars=["state"], 
                                          value_vars=["Amount released during FY 2019-20",
                                                      "Amount released during FY 2020-21",
                                                      "Amount released during FY 2021-22"], 
                                          var_name="Year", 
                                          value_name="Funding_Amount")

# Convert Year column to a proper format
state_funding_melted["Year"] = state_funding_melted["Year"].str.extract(r'(\d{4}-\d{2})')

# Convert Timestamp to datetime
data["Timestamp"] = pd.to_datetime(data["Timestamp"])

# Extract yearly average PM2.5 for Assam
assam_pm25 = data[data["state"] == "Assam"].groupby(data["Timestamp"].dt.year)["PM2.5"].mean().reset_index()

# Extract NCAP funding data for Assam
assam_funding = state_funding_melted[state_funding_melted["state"] == "Assam"][["Year", "Funding_Amount"]]

# Convert Year to numeric for merging
assam_funding["Year"] = assam_funding["Year"].str[:4].astype(int)

# Merge PM2.5 levels with funding data
assam_trends = assam_pm25.merge(assam_funding, left_on="Timestamp", right_on="Year", how="left").drop(columns="Year")

import matplotlib.ticker as mticker

# Create figure and primary y-axis (PM2.5)
fig, ax1 = plt.subplots(figsize=(8, 5))

# Plot PM2.5 levels (Red Line)
ax1.plot(assam_trends["Timestamp"], assam_trends["PM2.5"], marker="o", color="red", label="PM2.5 Levels")
ax1.set_xlabel("Year")
ax1.set_ylabel("PM2.5 Levels (µg/m³)", color="red")
ax1.tick_params(axis="y", labelcolor="red")

# Create secondary y-axis (Funding)
ax2 = ax1.twinx()
ax2.plot(assam_trends["Timestamp"], assam_trends["Funding_Amount"], marker="s", color="blue", label="NCAP Funding")
ax2.set_ylabel("NCAP Funding (Cr)", color="blue")
ax2.tick_params(axis="y", labelcolor="blue")

# Formatting and title
ax1.xaxis.set_major_locator(mticker.MaxNLocator(integer=True))
fig.suptitle("PM2.5 Levels vs. NCAP Funding in Assam", fontsize=14)
ax1.grid()

plt.show()
```

```python
#Q1) For the most polluted city in 2023 (highest average pollution in 2023), calculate the percentage of days with PM2.5 levels above the hazardous 
# threshold (300 µg/m³)[1 Mark]



# Filter data for 2023
data_2023 = data[data["Timestamp"].dt.year == 2023]

# Compute average PM2.5 per city
city_avg_pm25 = data_2023.groupby("city")["PM2.5"].mean()

# Identify the most polluted city
most_polluted_city = city_avg_pm25.idxmax()
highest_avg_pm25 = city_avg_pm25.max()

print(f"The most polluted city in 2023 is {most_polluted_city} with an average PM2.5 of {highest_avg_pm25:.2f} µg/m³.")

# Filter data for the most polluted city
city_data_2023 = data_2023[data_2023["city"] == most_polluted_city]

# Count total days & hazardous days (PM2.5 > 300)
total_days = city_data_2023["Timestamp"].nunique()
hazardous_days = city_data_2023[city_data_2023["PM2.5"] > 300]["Timestamp"].nunique()

# Compute percentage of hazardous days
hazardous_percentage = (hazardous_days / total_days) * 100

print(f"{most_polluted_city} had {hazardous_percentage:.2f}% of days with hazardous PM2.5 levels in 2023.")
```

```python
#Q2) Analyze the PM2.5 levels in Delhi during Diwali (October-November) and New Year compared to regular months over the 3 years (2018-2020). 
# Do you observe any significant pollution spikes during these festivals? If spikes are observed, Can you mark these spikes using arrows?[1.5 Mark]



# Ensure Timestamp column is in datetime format
data["Timestamp"] = pd.to_datetime(data["Timestamp"])

# Create an independent copy of the filtered dataset to avoid SettingWithCopyWarning
festival_data = data[(data["Timestamp"].dt.year >= 2018) & (data["Timestamp"].dt.year <= 2020)].copy()

# Define festival periods
diwali_months = [10, 11]  # October-November for Diwali
new_year_week = (festival_data["Timestamp"].dt.month == 1) & (festival_data["Timestamp"].dt.day <= 7)

# Assign Festival_Period safely using .loc[]
festival_data.loc[:, "Festival_Period"] = festival_data["Timestamp"].dt.month.isin(diwali_months) | new_year_week

# Compute average PM2.5 for festivals vs. non-festival periods
festival_comparison = festival_data.groupby("Festival_Period")["PM2.5"].mean()

print(f"PM2.5 during Festivals: {festival_comparison[True]:.2f} µg/m³")
print(f"PM2.5 during Regular Days: {festival_comparison[False]:.2f} µg/m³")

# Plot PM2.5 levels over time
plt.figure(figsize=(10, 5))
plt.plot(festival_data["Timestamp"], festival_data["PM2.5"], color="gray", label="PM2.5 Levels")

# Highlight festival periods
festival_days = festival_data[festival_data["Festival_Period"]]
plt.scatter(festival_days["Timestamp"], festival_days["PM2.5"], color="red", label="Festival Periods")

# Labels & legend
plt.xlabel("Year")
plt.ylabel("PM2.5 Levels (µg/m³)")
plt.title("PM2.5 Levels During Festivals (Diwali & New Year) vs. Regular Days")
plt.legend(loc="upper right")  # Set legend position manually
plt.grid()
plt.show()
```

insight 1: Which Cities Show the Fastest Decline in Air Quality?

```python
#Q1) Tell us something interesting from the dataset that we have not asked. Anything that intrigues you, anything that might be useful to the 
# policymakers. 


# Group data by year and city, then compute average PM2.5
city_pm25_trends = data.groupby([data["Timestamp"].dt.year, "city"])["PM2.5"].mean().reset_index()

# Compute PM2.5 change between first and last year for each city
pm25_change = city_pm25_trends.groupby("city")["PM2.5"].agg(["first", "last"])
pm25_change["Increase (%)"] = ((pm25_change["last"] - pm25_change["first"]) / pm25_change["first"]) * 100

# Find the top 5 cities with the highest increase in PM2.5
fastest_worsening_cities = pm25_change.nlargest(5, "Increase (%)")
print(fastest_worsening_cities)
```
